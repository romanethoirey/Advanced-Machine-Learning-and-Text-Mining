```python
from custom_lda import CustomLDA
import pandas as pd
from sklearn.decomposition import LatentDirichletAllocation, TruncatedSVD, NMF

data = pd.read_json('tweets.json')[0].tolist()
data = [str(d) for d in data]
```

```python
lda_model = LatentDirichletAllocation(max_iter=5, learning_method='online', learning_offset=50.,random_state=0)
c_lad = CustomLDA()
c_lad.build_model(data, _model=lda_model)
c_lad.save_model()
t = c_lad.document_topic_matrix()
print(t)
print(c_lad.show_topics())
```

```python
def apply_predict_topic(text):
    text = [text]
    infer_topic, topic, prob_scores = c_lad.predict_topic(text = text)
    return(infer_topic)
df = pd.DataFrame()
df['text'] = data
df["Topic_key_word"]= df['text'].apply(apply_predict_topic)
```
